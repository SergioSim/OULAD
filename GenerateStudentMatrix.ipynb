# Generate student matrix and export result to OULAD directory

```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
from tools.load_oulad import dataset_dict
import tools.filter_oulad as filter_oulad
```

```python
export_df = pd.DataFrame()
for i, line in dataset_dict['courses'].iterrows():
    oneCourse = filter_oulad.getOneCourse(dataset_dict, line.code_module, line.code_presentation)
    final_df = filter_oulad.restructure(oneCourse, 14)
    final_df["code_module"] = line.code_module
    final_df["code_presentation"] = line.code_presentation
    export_df = export_df.append(final_df)
export_df
```

```python
export_df.to_csv("./OULAD/student_matrix.csv")
```

```python
filter_oulad.cleanAndMap(export_df)
export_df
```
